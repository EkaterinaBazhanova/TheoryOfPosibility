```python
# Определим необходимые функции

from math import factorial

def arrangements(n, k):
    return int(factorial(n) / factorial(n - k))

def combinations(n, k):
    return int(factorial(n) / (factorial(k) * factorial(n - k)))
```

```python
# Расписание на понедельник -- размещение из 10 по 4 (важен состав и порядок)
print(f'Количество вариантов расписания на понедельник: {arrangements(10, 4)}')
```

**№2.** Из колоды в 52 карты вынимают случайным образом 4 карты. Найти число исходов, соответствующих тому, что был вытянут хотя бы один туз.

```python
# Разделим все карты на 2 типа: тузы и остальные.
# Пусть A_i -- из 4-х вытянутых карт i тузoв, i = 1, 2, 3, 4.
# Тогда A -- из 4-х вытянутых карт хотя бы один туз и A = A_1 + A_2 + A_3 + A_4.

# Число исходов, соответствующих событию A_i = выбор i тузов из 4 и выбор (4-i) карт из 48 оставшихся:
# C_{4}^{i}*C_{48}^{4-i}

P = 0
for i in range(1,5):
    P_i = combinations(4, i)*combinations(48, 4-i)
    P = P + P_i

print(f'Число исходов, соответствующих тому, что был вытянут хотя бы один туз: {P}')
```

**№5.** Бросается игральная кость. Пусть событие `A` - появление чётного числа, событие `B` - появление числа больше трёх. Являются ли эти события независимыми?

**№6.** Допустим, имеется некоторая очень редкая болезнь (поражает 0.1 % населения). Вы приходите к врачу, вам делают тест на эту болезнь, и тест оказывается положительным. Врач говорит вам, что этот тест верно выявляет 99 % больных этой болезнью и всего лишь в 1 % случаев даёт ложный положительный ответ.

Вопрос: какова вероятность, что вы действительно больны ей?

Подсказка: вновь используйте формулу Байеса с раскрытием знаменателя с помощью формулы полной вероятности.

```python
# Пусть H_1 -- человек болен
# P_1 = Р(Н_1) = 0.1 / 100
P_H1 = 0.1 / 100

# Пусть H_2 -- человек здоров
# P_2 = Р(Н_2) = 1 - Р(Н_1)
P_H2 = 1 - P_H1

# Пусть A -- тест положительный, тогда
# P_AH1 = P(A|H_1) = 99/100
P_AH1 = 99/100

# P_AH2 = P(A|H_2) = 1/100
P_AH2 = 1/100

print('Событие A -- тест положительный')
print('Гипотеза H_1 -- человек болен')
print('Гипотеза H_2 -- человек здоров \n')
print(f'P(H_1) = {P_H1}', '     ', f'P(A|H_1) = {P_AH1}')
print(f'P(H_2) = {P_H2}', '     ', f'P(A|H_2) = {P_AH2} \n')

#Тогда по формуле полной вероятности
#P_A = P(A) = P(H_1)*P(A|H_1) + P(H_2)*P(A|H_2)
P_A = P_H1 * P_AH1 + P_H2 * P_AH2

print(f'P(A) = P(H_1)*P(A|H_1) + P(H_2)*P(A|H_2) = {P_A} \n')

# и по формуле Байеса:
# P(H_1|A) = P(H_1)*P(A|H_1) / P(A)
P_H1A = P_H1 * P_AH1 / P_A

print(f'Вероятность болезни при положительном тесте: P(H_1|A) = P(H_1)*P(A|H_1) / P(A) = {P_H1A}')
```
